```python
# Importing necessary libraries for data manipulation
import numpy as np
import pandas as pd
```

```python
# Load dataset from source directory into a dataframe
df = pd.read_csv(r"C:\Users\chris\OneDrive\Documents\paschal\END_OF_THE_YEAR_ANALYSIS\TDI Financial Consumer Complaints\Financial Consumer Complaints.csv")
```

```python
# Display data summary for familiarization 
df.head(5)
```

```python
# Rename dataframe df as 'Cumulus financial' cf 
cf = df
cf.head(5)
```

```python
# Remove duplicates 
cf = cf.drop_duplicates()

cf.head()
```

```python
# Checking dataset information
cf.info()
```

```python
# Renamed the "Date Submitted" column for clarity and consistency.
cf.rename(columns = {"Date Sumbited": "Date submitted"}, inplace=True)
```

```python
# Reformat Date submitted column from object to datetime 
cf['Date submitted'] = pd.to_datetime(cf['Date submitted'], errors='coerce')
```

```python
# Reformat Date received column from object to datetime 
cf['Date Received'] = pd.to_datetime(cf['Date Received'], errors='coerce')
```

```python
# Verifying formatting changes
cf.info()
```

```python
# Visualizing reformatted dataframe 
cf.head(4)
```

```python
# Removed the "Company Public Response" column, which was deemed non-essential to the analysis.
cf.drop(columns = ['Company public response'], inplace=True) 
```

```python
# Updated the "Issue" column to combine "Advertising" with "Advertising and Marketing" category for better classification.
cf['Issue'] = cf['Issue'].replace({'Advertising': 'Advertising and marketing'})
```

```python
# To verify changes reflected in the Issue column 
unique_issues = pd.DataFrame(cf['Issue'].unique(), columns=['Unique Issues'])

# Sorted alphabetically
unique_issues = unique_issues.sort_values(by='Unique Issues').reset_index(drop=True)

unique_issues.head(10)
```

```python
# Consolidated "Credit or Prepaid Card" under the "Product" column into a single "Credit Card" category.
cf['Product'] = cf['Product'].replace({'Credit or Prepaid Card': 'Credit Card'})
```

```python
#  Replaced ambiguous entries (“) in the "Sub-Product" column with "Not Specified" to ensure completeness.

cf['Sub-product'] = cf['Sub-product'].replace({'""': 'Not Specified'}, regex=True)
```

```python
#  Replaced ambiguous entries (“) in the "Sub-Issue" column with "Not Specified" to ensure completeness.

cf['Sub-issue'] = cf['Sub-issue'].replace({'""': 'Not Specified'}, regex=True)
```

```python
# Verifying changes 
cf.head(10)
```

```python
# Dropped the "Tags" column due to a high volume of missing values, rendering it unsuitable for analysis.
cf.drop(columns = ['Tags'], inplace=True) 
```

```python
# Removed the "Consumer Consent Provided" column due to limited relevance.
cf.drop(columns = ['Consumer consent provided?'], inplace=True) 
```

```python
# Excluded the "Zip Code" field from the dataset, as it had a substantial proportion of missing data, which hindered its utility in analysis
cf.drop(columns = ['ZIP code'], inplace=True) 
```

```python
# Printing cleaned dataset for final inspection 
cf.head(20)
```

```python
# Export the clean data to local storage
cf.to_excel('Financial Consumer Complaints clean.xlsx')
```
